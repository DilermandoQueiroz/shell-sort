```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

# Carregamento dos dados
df = pd.read_csv('Source/sorting_results.csv')

# Visualização inicial dos dados
df.head()
```

# Análise de Desempenho de Algoritmos de Ordenação

## Preparação do Ambiente

Neste notebook, vamos analisar o desempenho de diferentes algoritmos de ordenação em diversos cenários. Os dados foram coletados para diferentes tamanhos de arranjos e tipos de ordenação.

### Importação de Bibliotecas e Carregamento dos Dados


```python
# Função para criar gráficos comparativos com escala logarítmica
def plot_log_comparison(metric, title, ylabel):
    plt.figure(figsize=(14, 8))
    sns.barplot(x='Size', y=metric, hue='Algorithm', data=df)
    plt.yscale('log')
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel('Tamanho do Arranjo')
    plt.legend(title='Algoritmo', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

```

## Análise Comparativa Inicial

Nesta seção, comparamos os algoritmos de ordenação com base no número de comparações, número de movimentações e tempo de execução total usando uma escala logarítmica para melhor visualização.

### Comparação de Número de Comparações)

```python
# Número de comparações de chaves com escala logarítmica
plot_log_comparison('Comparisons', 'Comparação de Número de Comparações entre Algoritmos (Escala Logarítmica)', 'Número de Comparações (log)')
```

### Comparação de Número de Movimentações

```python
# Número de movimentações de registros com escala logarítmica
plot_log_comparison('Movements', 'Comparação de Número de Movimentações entre Algoritmos (Escala Logarítmica)', 'Número de Movimentações (log)')
```

### Comparação de Tempo de Execução

```python
# Tempo total gasto para ordenação com escala logarítmica
plot_log_comparison('Time(us)', 'Comparação de Tempo de Execução entre Algoritmos (Escala Logarítmica)', 'Tempo de Execução (us) (log)')
```

## Análise Detalhada por Tipo de Arranjo

```python
# Bloco 3: Análise Detalhada por Tipo de Arranjo

# Função para criar gráficos comparativos por tipo de arranjo com escala logarítmica
def plot_log_comparison_by_type(array_type, metric, title, ylabel):
    plt.figure(figsize=(14, 8))
    subset = df[df['ArrayType'] == array_type]
    sns.barplot(x='Size', y=metric, hue='Algorithm', data=subset)
    plt.yscale('log')
    plt.title(f'{title} ({array_type})')
    plt.ylabel(ylabel)
    plt.xlabel('Tamanho do Arranjo')
    plt.legend(title='Algoritmo', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()


```

### Arranjos Ordenados

```python
plot_log_comparison_by_type('Sorted', 'Comparisons', 'Comparação de Número de Comparações entre Algoritmos', 'Número de Comparações (log)')
plot_log_comparison_by_type('Sorted', 'Movements', 'Comparação de Número de Movimentações entre Algoritmos', 'Número de Movimentações (log)')
plot_log_comparison_by_type('Sorted', 'Time(us)', 'Comparação de Tempo de Execução entre Algoritmos', 'Tempo de Execução (us) (log)')


# Tipos de arranjos
array_types = ['Sorted', 'Reversed', 'AlmostSorted', 'Random']
```

### Arranjos Inversamente Ordenados

```python
plot_log_comparison_by_type('Reversed', 'Comparisons', 'Comparação de Número de Comparações entre Algoritmos', 'Número de Comparações (log)')
plot_log_comparison_by_type('Reversed', 'Movements', 'Comparação de Número de Movimentações entre Algoritmos', 'Número de Movimentações (log)')
plot_log_comparison_by_type('Reversed', 'Time(us)', 'Comparação de Tempo de Execução entre Algoritmos', 'Tempo de Execução (us) (log)')


# Tipos de arranjos

array_types = ['Sorted', 'Reversed', 'AlmostSorted', 'Random']
```

### Arranjos Quase Ordenados

```python
plot_log_comparison_by_type('AlmostSorted', 'Comparisons', 'Comparação de Número de Comparações entre Algoritmos', 'Número de Comparações (log)')
plot_log_comparison_by_type('AlmostSorted', 'Movements', 'Comparação de Número de Movimentações entre Algoritmos', 'Número de Movimentações (log)')
plot_log_comparison_by_type('AlmostSorted', 'Time(us)', 'Comparação de Tempo de Execução entre Algoritmos', 'Tempo de Execução (us) (log)')

```

### Arranjos Aleatórios

```python
plot_log_comparison_by_type('Random', 'Comparisons', 'Comparação de Número de Comparações entre Algoritmos', 'Número de Comparações (log)')
plot_log_comparison_by_type('Random', 'Movements', 'Comparação de Número de Movimentações entre Algoritmos', 'Número de Movimentações (log)')
plot_log_comparison_by_type('Random', 'Time(us)', 'Comparação de Tempo de Execução entre Algoritmos', 'Tempo de Execução (us) (log)')
```

## Insights e Discussões

### Eficiência Assintótica vs. Empírica

Os algoritmos com complexidade \(O(n^2)\), como InsertionSort e BubbleSort, mostraram ser menos eficientes em prática para grandes volumes de dados, conforme esperado. Em contrapartida, os algoritmos com complexidade \(O(n \log n)\), como HeapSort, MergeSort e QuickSort, apresentaram um desempenho empírico superior, refletindo suas análises assintóticas.

### Impacto do Tipo de Arranjo

- **InsertionSort:** Apresenta um desempenho significativamente melhor em arranjos quase ordenados e pior em arranjos inversamente ordenados.
- **BubbleSort:** Consistentemente ineficiente para todos os tipos de arranjo e tamanhos maiores.
- **HeapSort, MergeSort e QuickSort:** Mantêm eficiência independente do tipo de arranjo, embora o QuickSort possa ter desempenho variável dependendo da escolha do pivô.

### Escalabilidade

Os gráficos em escala logarítmica destacam a superioridade dos algoritmos \(O(n \log n)\) para grandes volumes de dados, mantendo tempos de execução mais baixos em comparação com os algoritmos \(O(n^2)\).

Esta análise fornece uma visão abrangente do desempenho dos algoritmos de ordenação em diferentes cenários, permitindo uma compreensão mais profunda de suas eficiências e comportamentos. Estes insights podem guiar a escolha do algoritmo mais adequado para diferentes contextos e tamanhos de dados.


## Análise Detalhada do Shell Sort

O Shell Sort é um algoritmo de ordenação que melhora o desempenho do Insertion Sort ao permitir comparações e trocas de elementos que estão distantes entre si. Ele usa uma sequência de incrementos (gaps) que são reduzidos até que o gap final seja 1, momento em que o algoritmo se comporta como um Insertion Sort.

### Insights sobre o Shell Sort

O Shell Sort apresenta um número de comparações e movimentações menor do que o Insertion Sort em arranjos maiores, devido à sua estratégia de gap. Em termos de tempo de execução, o Shell Sort é significativamente mais eficiente do que os algoritmos \(O(n^2)\), especialmente em arranjos grandes.

### REESCREVER!!!! Comportamento Esperado vs. Observado

**REESCREVER!!!! Comportamento Esperado:**
- O Shell Sort deve reduzir o número de comparações e movimentações em comparação com o Insertion Sort, especialmente para arranjos maiores e não ordenados.
- O desempenho deve ser melhor do que o Bubble Sort e, em alguns casos, pode ser comparável a algoritmos \(O(n \log n)\).

** REESCREVER!!!!Comportamento Observado:**
- Os dados confirmam que o Shell Sort é mais eficiente do que o Insertion Sort e Bubble Sort em termos de comparações, movimentações e tempo de execução.
- Em arranjos grandes, o Shell Sort mantém um desempenho eficiente, embora não tão bom quanto os algoritmos \(O(n \log n)\) como MergeSort e QuickSort.

Esta análise detalhada do Shell Sort confirma sua eficiência em comparação com outros algoritmos \(O(n^2)\) e destaca sua utilidade em cenários onde é necessário um compromisso entre simplicidade e desempenho.
