```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv('sorting_results.csv')
results
```

```python
results_group = results.groupby(['Algorithm', 'ArrayType', 'Size']).mean()
```

# Analise de computação de chaves

# Analise de tempo de computação

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=results_group, x='Size', y='Time(us)', hue='Algorithm', marker='o')
plt.yscale('log')
plt.xscale('log')
plt.title(f'Tempo de Processamento por Tamanho do Vetor (Escala Logarítmica)')
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Tempo de Processamento (µs)')
plt.legend(title='Algoritmo')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
for array_type in results['ArrayType'].unique():
    plt.figure(figsize=(14, 8))
    sns.lineplot(data=results[results['ArrayType'] == array_type], x='Size', y='Time(us)', hue='Algorithm', marker='o')
    plt.yscale('log')
    plt.xscale('log')
    plt.title(f'Tempo de Processamento por Tamanho do Vetor {array_type} (Escala Logarítmica)')
    plt.xlabel('Tamanho do Vetor')
    plt.ylabel('Tempo de Processamento (µs)')
    plt.legend(title='Algoritmo')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()
```
